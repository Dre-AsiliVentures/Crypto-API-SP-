```python
import os
import pandas as pd
import talib
import numpy
from binance import Client

```

```python
class Binance_DataFetch():
    def __init___():
        import pandas as pd
        import numpy as np
        import binance.Client
        import os
        print("Done importing all libraries")
    def client_setup():
        api_key = os.environ.get('BINANCE_TEST_API')
        api_secret = os.environ.get('BINANCE_TEST_SECRET')
        client=Client(api_key, api_secret)
    def client_dataFetch(self,ticker_symbol,period_interval,lookback_period):
        print("Now fetching the client data")
        data=client.get_historical_klines(ticker_symbol,period_interval,lookback_period)
        frame=pd.DataFrame(data)
        frame=frame.iloc[:,:6] # All rows and column upto 6
        # Namig the columns
        frame.columns=['Time','Open','High','Low','Close','Volume']
        # Name the rows
        frame=frame.set_index('Time') # Note this is in ms
        # Convert this Time to readable form since this is time from 1970s
        frame.index=pd.to_datetime(frame.index,unit='ms')
        frame=frame.astype(float) # Convert values from strings to floatn
        return frame

"""Example usage:
a=Binance_DataFetch()
data=a.client_dataFetch("BNBBUSD", Client.KLINE_INTERVAL_1DAY, "2000 day ago UTC+3")
# #https://python-binance.readthedocs.io/en/latest/market_data.html?highlight=kline%20limit#id6
"""
```
